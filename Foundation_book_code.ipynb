```python
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
# making sure that my code matches the book

# data vs simple linear regression

xx = [4, 5, 6, 7, 8, 9]
y_real = [20, 25, 29, 31, 32, 32]

np_arr_x = np.array(xx)
np_arr_y = np.array(y_real)

y_dash = -13.2 + 11*np_arr_x - 0.66*(np_arr_x**2)
np_arr_y_dash = y_dash

simple_regressor = LinearRegression()
simple_reg_line = simple_regressor.fit(np_arr_x.reshape(-1,1), np_arr_y)
prediction = simple_regressor.predict(np_arr_x.reshape(-1,1))

plt.scatter(np_arr_x, np_arr_y)
# plt.plot(np_arr_x, np_arr_y_dash)
plt.plot(np_arr_x, prediction)
plt.title('Fitted simple Regression line')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# we can see clearly that there is a non-linear relationship lift after removing the linear relationship
# that's a strong indicator that there is a non-linear relationship

plt.scatter(prediction, np_arr_y - prediction)
plt.title('Risidual plot')
plt.xlabel('predicted y')
plt.ylabel('Risiduals (y - pred_y)')
```

```python
# data vs polynomial regression (non-linear)

xx = [4, 5, 6, 7, 8, 9]
y_real = [20, 25, 29, 31, 32, 32]

np_arr_x = np.array(xx)
np_arr_y = np.array(y_real)

y_dash = -13.2 + 11*np_arr_x - 0.66*(np_arr_x**2)

plt.scatter(np_arr_x, np_arr_y)
plt.plot(np_arr_x, y_dash)
plt.title('Fitted plynomial Regression line')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# no trend after removing the non-linear relationship

plt.scatter(y_dash, np_arr_y - y_dash)
plt.title('Risidual plot')
plt.xlabel('predicted y')
plt.ylabel('Risiduals (y - pred_y)')
plt.xlim(20, 34)
plt.ylim(-3, 3)
```
